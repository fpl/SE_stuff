# 1) Dealing with Data Type ( grade 20%). 

Search the right gdal command to change the data type to Byte for these 2 files. <br />

/home/user/SE_data/exercise/geodata/LST/LST_MOYDmax_month4.tif <br />
/home/user/SE_data/exercise/geodata/dem/GMTED2010.tif <br />



```python
%%bash
gdal_translate -co COMPRESS=DEFLATE -co ZLEVEL=9  -ot Byte ~/SE_data/exercise/geodata/LST/LST_MOYDmax_month4.tif ~/SE_mydata/LST_MOYDmax_month4_byte.tif
gdal_translate -co COMPRESS=DEFLATE -co ZLEVEL=9  -ot Byte ~/SE_data/exercise/geodata/dem/GMTED2010.tif ~/SE_mydata/GMTED2010_byte.tif
```

Calculate min and max to see the effect of the change data type

```python
%%bash
gdalinfo -mm  ~/SE_data/exercise/geodata/LST/LST_MOYDmax_month4.tif  | grep Computed
gdalinfo -mm  ~/SE_mydata/LST_MOYDmax_month4_byte.tif | grep Computed
```

```python
%%bash
gdalinfo -mm  ~/SE_data/exercise/geodata/dem/GMTED2010.tif  | grep Computed
gdalinfo -mm  ~/SE_mydata/GMTED2010_byte.tif | grep Computed
```

## A) Create a loop that extract each raster, save as a compressed tif.

```python
%%bash 
cd ~/SE_mydata/terraclimate/
## gdalinfo to understand the data 
gdalinfo TerraClimate_tmax_2017.nc
## 
```

```python
%%bash
## metadata of the subset NETCDF:"/home/user/SE_mydata/terraclimate/TerraClimate_tmax_2017.nc":tmax
cd /home/user/SE_mydata/terraclimate/
gdalinfo NETCDF:"TerraClimate_tmax_2017.nc":tmax
```

```python
%%bash
cd /home/user/SE_mydata/terraclimate/
for MM in {1..12} ; do
### something is missing .....
gdal_translate -co COMPRESS=DEFLATE -co ZLEVEL=9  NETCDF:"TerraClimate_tmax_2017.nc":tmax  tmax_2017_${MM}.tif
done
```

## B) Which is the pixel resolution in degree unit (arc-second or in arc-minutes) and in meter (approximation at the equator is fine)?

We know that 0.008333333 degree = 30''  = ~1km at the equator  <br />
Therefore 0.041666666666667 / 0.008333333 = 5 , so 5 * 30'' = 150'' = 2' 30'' = ~5km (actually is more close to 4km)  <br />

## C) Build the loop to download and extract 5 years. You can find the files under the same directory (https://climate.northwestknowledge.net/TERRACLIMATE-DATA/ ) this count as bonus grade +10% ). 

```python
%%bash
cd /home/user/SE_mydata/terraclimate
# let's do it only 2 years
for YYYY in {2000..2002} ; do 
wget https://climate.northwestknowledge.net/TERRACLIMATE-DATA/TerraClimate_tmax_${YYYY}.nc
for MM in {1..12} ; do
gdal_translate -co COMPRESS=DEFLATE -co ZLEVEL=9 NETCDF:"TerraClimate_tmax_${YYYY}.nc":tmax tmax_${YYYY}_${MM}.tif
done
done
```
